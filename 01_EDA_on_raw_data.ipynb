Temporal consistency

```python
"""
check relations between:

founded_at ≤ first_funding_at
first_funding_at ≤ last_funding_at
closed_at ≥ founded_at (if the closed_at is not NaN)
no funding rounds after closed_at


"""
```

```python
# checking if founded_at ≤ first_funding_at

#make sure column values are in the date format
startups["founded_at"] = pd.to_datetime(startups["founded_at"], errors="coerce")
startups["first_funding_at"] = pd.to_datetime(startups["first_funding_at"], errors="coerce")

non_null = startups["first_funding_at"].notna() & startups["founded_at"].notna()
temporal_validation = startups["first_funding_at"] > startups["founded_at"]

violations = startups[non_null & temporal_validation]
violations
```

The majority of records with non-missing values satisfy the condition first_funding_at > founded_at, suggesting that funding timelines are largly temporally coherent.

```python
#checking if first_funding_at ≤ last_funding_at

startups["first_funding_at"] = pd.to_datetime(startups["first_funding_at"], errors="coerce")
startups["last_funding_at"] = pd.to_datetime(startups["last_funding_at"], errors="coerce")

non_null = startups["first_funding_at"].notna() & startups["last_funding_at"].notna()

temporal_validation2 = startups["last_funding_at"] >= startups["first_funding_at"]

valid_records = startups[non_null & temporal_validation2]
valid_records
```

The resulting table contains the same number of rows as the source dataset, indicating that all records with non-missing funding dates satisfy the condition last_funding_at >= first_funding_at.
