# Missing data analysis

```python
startups.isna().sum()

no_of_missing_values = startups.isnull().sum()
percent_missing = round(startups.isnull().sum() * 100 / len(startups), 2)
missing_values_df = pd.DataFrame({
                                 'no_of_missing_values': no_of_missing_values,
                                 'percent_missing': percent_missing})

missing_values_df
```

Observations:



*   493 NaNs for Unnamed: 6 - contains a high proportion of missing values and appears to be CSV atrifact. This column does not carry any informative value and will be considered for the removal in the data cleaning stage.
*   588 NaNs for closed_at - the missing entries most likely indicate that these starups were still operating at the time the dataset was recorded.
*   152 NaNs for age_first_milestone_year - these starups likely didn't reach their first milestone. Therefore this column, as the previous one, is informative and won't be considered for removal.
*   152 age_last_milestone_year - same as the age_first_milestone_year.
*   state_code.1 - one missing value, probably due to minor data inconsistency.
